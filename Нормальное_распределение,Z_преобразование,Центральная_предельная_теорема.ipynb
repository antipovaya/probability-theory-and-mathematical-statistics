**1. Нормальное распределение (Гауссово распределение)**  

Нормальное распределение — одно из важнейших распределений в статистике, описывающее многие природные явления. Это распределение вероятностей **непрерывной** случайной величины Х, плотность вероятности которой подчиняется формуле:

Функция плотности вероятности:  

**f(x) = (1/(σ√(2π))) * e^(-(x-μ)²/(2σ²))**

где:

μ - математическое ожидание (центр распределения)

σ - стандартное отклонение (мера разброса)

π ≈ 3.14159, e ≈ 2.71828

**Правило трех сигм:**

Симметрично относительно среднего μ

68,26% данных в пределах μ ± σ

95,44% данных в пределах μ ± 2σ

99.72% данных в пределах μ ± 3σ

**2. Z-преобразование (стандартное нормальное распределение)**  

Z-преобразование позволяет преобразовать любое нормальное распределение в стандартное нормальное распределение с параметрами μ=0, σ=1.

Формула:
z = (x - μ) / σ

где:

x - исходное значение

μ - среднее генеральной совокупности

σ - стандартное отклонение генеральной совокупности

Смысл: Z-score показывает, на сколько стандартных отклонений значение отклоняется от среднего.  


**3. Работа с таблицами Z-значений**  

Таблица Z-значений (таблица стандартного нормального распределения) показывает вероятность того, что случайная величина примет значение меньше или равное z.

Как пользоваться таблицей:

Найдите первую цифру z в левом столбце

Найдите вторую цифру в верхней строке

На пересечении - искомая вероятность P(Z ≤ z)

Пример: Найти P(Z ≤ 1.25)

В левом столбце находим 1.2

В верхней строке находим 0.05

На пересечении: 0.8944

Для отрицательных z: Используем свойство симметрии:
P(Z ≤ -z) = 1 - P(Z ≤ z)

**4. Центральная предельная теорема (ЦПТ)**  

Формулировка: При достаточно большом объеме выборки (n ≥ 30) выборочное среднее имеет нормальное распределение независимо от распределения генеральной совокупности.

Ключевые положения:  


Среднее выборочных средних равно среднему генеральной совокупности: μₓ̄ = μ

Стандартная ошибка среднего: σₓ̄ = σ/√n

Для выборочных средних: z = (x̄ - μ) / (σ/√n)

```python
# Установка необходимых библиотек (если нужно)
!pip install scipy numpy matplotlib

# Импорт
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

**1. Генерация нормально распределенных данных**

```python
# Параметры распределения
mu = 50      # Среднее
sigma = 10   # Стандартное отклонение
size = 1000  # Размер выборки

# Генерация данных
# Эта строка генерирует массив из `size` случайных чисел, распределенных по нормальному закону с средним `mu` и стандартным отклонением `sigma`.
data = np.random.normal(mu, sigma, size)

# Визуализация
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')

# - data: данные для построения гистограммы.
# - bins=30: количество столбцов гистограммы.
# - density=True: нормировка гистограммы, чтобы площадь под ней была равна 1 (т.е. показывает плотность вероятности).
# - alpha=0.7: прозрачность столбцов.
# - color='blue': цвет столбцов.
# - edgecolor='black': цвет границ столбцов.

plt.title('Нормальное распределение N(50, 10)')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.grid(True)

# Добавляем теоретическую кривую
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma), 'r-', lw=2, label='Теоретическая кривая')
# x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100) - создает массив x из 100 точек в диапазоне от mu-4*sigma до mu+4*sigma.
# plt.plot(x, stats.norm.pdf(x, mu, sigma), 'r-', lw=2, label='Теоретическая кривая') - строит теоретическую кривую плотности нормального распределения.
# - stats.norm.pdf(x, mu, sigma) вычисляет значение плотности в точках x для нормального распределения с параметрами mu и sigma.
# - 'r-' означает красную сплошную линию.
# - lw=2 - толщина линии.
# - label='Теоретическая кривая' - метка для легенды.
plt.legend()
plt.show()
```

**2. Z-стандартизация и работа с таблицами**

```python
# Пример данных
scores = np.array([72, 85, 90, 65, 78, 88, 92, 60, 75, 81])
mu = 75  # Предполагаемое среднее популяции
sigma = 10  # Предполагаемое стандартное отклонение популяции

# Z-стандартизация
z_scores = (scores - mu) / sigma
print("Z-скоры:", z_scores)

# Находим вероятности используя scipy (вместо таблиц)
probabilities = stats.norm.cdf(z_scores)
print("Вероятности P(Z ≤ z):", probabilities)

# Обратная задача: из вероятности в z-значение
target_prob = 0.95
z_value = stats.norm.ppf(target_prob)
print(f"Z-значение для вероятности {target_prob}: {z_value:.4f}")

# Обратная задача: из z-значения в исходное значение
original_value = z_value * sigma + mu
print(f"Исходное значение для z={z_value:.4f}: {original_value:.2f}")
```

**3. Центральная предельная теорема на практике**


```python
# Демонстрация ЦПТ
np.random.seed(42)  # Для воспроизводимости

# Генерируем данные из ненормального распределения (экспоненциального)
population = np.random.exponential(2, 10000)

# Берем 1000 выборок по 30 элементов
sample_means = []
n_samples = 1000
sample_size = 30

for _ in range(n_samples):
    sample = np.random.choice(population, sample_size, replace=False)
    sample_means.append(np.mean(sample))

# Визуализируем
plt.figure(figsize=(12, 5))

# Исходное распределение
plt.subplot(1, 2, 1)
plt.hist(population, bins=30, density=True, alpha=0.7, color='green', edgecolor='black')
plt.title('Исходное экспоненциальное распределение')
plt.xlabel('Значение')
plt.ylabel('Плотность')

# Распределение выборочных средних
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='orange', edgecolor='black')
plt.title('Распределение выборочных средних (n=30)')
plt.xlabel('Среднее значение')
plt.ylabel('Плотность')

# Добавляем нормальную кривую для сравнения
x = np.linspace(min(sample_means), max(sample_means), 100)
mean_means = np.mean(sample_means)
std_means = np.std(sample_means, ddof=1)
plt.plot(x, stats.norm.pdf(x, mean_means, std_means), 'r-', lw=2)

plt.tight_layout()
plt.show()

print(f"Среднее исходного распределения: {np.mean(population):.4f}")
print(f"Среднее выборочных средних: {mean_means:.4f}")
print(f"Стандартная ошибка: {std_means:.4f}")
print(f"Теоретическая стандартная ошибка: {np.std(population)/np.sqrt(sample_size):.4f}")
```

Задача 1: Рост взрослых мужчин распределен нормально с μ = 175 см, σ = 7 см. Какова вероятность, что случайно выбранный мужчина будет ростом:  

а) Меньше 180 см?  

б) Больше 190 см?  

в) От 170 до 185 см?  


z = (X - μ) / σ

```python
# Решение
mu = 175
sigma = 7

# а) Меньше 180 см
z_a = (180 - mu) / sigma
print(f"z-наблюдаемое = {z_a:.4f}")
print("Ниже по таблице z-значений ищем вероятность, что случайная величина будет ниже 180 (иначе найдем долю значений СВ, которые лежат ниже отметки 180) - 76%")
# Далее проверим по Функции распределения. Эта функция возвращает вероятность того, что случайная величина примет значение меньше или равное x.
prob_a = stats.norm.cdf(z_a)
print(f"а) Вероятность роста < 180 см: {prob_a:.4f} или {prob_a*100:.2f}%")
```

```python
# б) Больше 190 см
z_b = (190 - mu) / sigma
print(f"z-наблюдаемое = {z_b:.4f}")
print("Ниже по таблице z-значений ищем вероятность, что случайная величина будет больше 190 (иначе найдем долю значений СВ, которые лежат выше отметки 190), т.е. 1 - z(2,14) - 1,61%")
# Далее проверим по Функции распределения. Эта функция возвращает вероятность того, что случайная величина примет значение меньше или равное x.
# Помним это, поэтому отнимаем единицу:
prob_b = 1 - stats.norm.cdf(z_b)
print(f"б) Вероятность роста > 190 см: {prob_b:.4f} или {prob_b*100:.2f}%")

```

```python
# в) От 170 до 185 см
z_c1 = (170 - mu) / sigma
z_c2 = (185 - mu) / sigma
print(f"z-наблюдаемое1 = {z_c1:.4f}")
print(f"z-наблюдаемое2 = {z_c2:.4f}")
print("Ниже по таблице z-значений ищем вероятность, что случайная величина будет от 170 до 185 см, т.е. z(1,4286) - z(-0,7143)")
# Далее проверим по Функции распределения. Эта функция возвращает вероятность того, что случайная величина примет значение меньше или равное x.
# Помним это, поэтому отнимаем:
prob_c = stats.norm.cdf(z_c2) - stats.norm.cdf(z_c1)
print(f"в) Вероятность роста между 170 и 185 см: {prob_c:.4f} или {prob_c*100:.2f}%")
```

Задача 2: Счеты IQ имеют нормальное распределение с μ = 100, σ = 15. Какой процент населения имеет IQ:  

а) Выше 125?  

б) Между 90 и 110?  

в) Найдите IQ, который превышают только 5% населения.  


```python
def Z(score, mean, std_dev):
  z_score = (score - mean) / std_dev
  return z_score
```

```python
mean_iq = 100
std_dev_iq = 15

# а) Выше 125?
z_iq_a = Z(125, 100, 15)
print(f"z-наблюдаемое = {z_iq_a:.4f}")
```

image.png

```python
# Для того, чтобы найти вероятность того, что СВ будет выше 125 нужно от 1 отнять z(1,6667) = 1 - 0,95154 = 0,04846
# Далее проверим по Функции распределения. Эта функция возвращает вероятность того, что случайная величина примет значение меньше или равное x.
# Помним это, поэтому отнимаем единицу:
prob_iq_a = 1 - stats.norm.cdf(z_iq_a)
print(f" IQ: а) Выше 125 имеет {prob_iq_a * 100 :.2f}% населения")
```

```python
# б) Между 90 и 110?
z_iq_b1 = Z(90, mean_iq, std_dev_iq)
z_iq_b2 = Z(110, mean_iq, std_dev_iq)
print(f"z-наблюдаемое1 = {z_iq_b1:.4f}")
print(f"z-наблюдаемое2 = {z_iq_b2:.4f}")
print("Ниже по таблице z-значений ищем вероятность, что случайная величина будет от 90 до 110, т.е. z(0,6667) - z(-0,6667)")
# Далее проверим по Функции распределения. Эта функция возвращает вероятность того, что случайная величина примет значение меньше или равное x.
# Помним это, поэтому отнимаем:
prob_iq_b = stats.norm.cdf(z_iq_b2) - stats.norm.cdf(z_iq_b1)
print(f"IQ: б) Между 90 и 110 имеет {prob_iq_b * 100:.2f}%")
```

```python
# в) IQ, который превышают только 5% населения
z_c_iq = stats.norm.ppf(0.95)  # 95-й процентиль
# Значение также можем посмотреть в таблице (ищем вероятность = 0,95, ей соответстует z = 1,65)
iq_value = z_c_iq * std_dev_iq + mean_iq
print(f"в) IQ, превышаемый только 5% населения: {iq_value:.2f}")
```
