**Подробное объяснение функции np.linspace()**  

Функция np.linspace() — одна из важнейших функций в NumPy для создания последовательностей чисел. Давайте разберем ее подробно.


**Синтаксис**

```python
# np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)
```

**Параметры**

start (обязательный): Начальное значение последовательности

stop (обязательный): Конечное значение последовательности

num (опциональный): Количество элементов в последовательности (по умолчанию 50)

endpoint (опциональный): Если True (по умолчанию), включает конечное значение stop в последовательность

retstep (опциональный): Если True, возвращает кортеж (массив, шаг)

dtype (опциональный): Тип данных элементов массива

axis (опциональный): Ось, вдоль которой создаются значения (для многомерных массивов)

**Что возвращает функция**

Функция возвращает одномерный массив с равномерно распределенными значениями из указанного интервала.

Примеры и объяснение  

**Базовый пример**

```python
import numpy as np

# Создает 5 чисел от 0 до 1 (включительно)
linspace = np.linspace(0, 1, 5)
print("linspace:", linspace)  # [0.0, 0.25, 0.5, 0.75, 1.0]
```

**Как рассчитываются значения**  

Значения рассчитываются по формуле:  

значение = start + i * ((stop - start) / (num - 1))  
для i от 0 до num-1

Для нашего примера:

start = 0, stop = 1, num = 5

Шаг = (1 - 0) / (5 - 1) = 0.25

Значения: 0 + 0*0.25, 0 + 1*0.25, 0 + 2*0.25, 0 + 3*0.25, 0 + 4*0.25

Результат: [0.0, 0.25, 0.5, 0.75, 1.0]

**Различные варианты использования**

```python
# Без включения конечной точки
linspace_no_end = np.linspace(0, 1, 5, endpoint=False)
print("Без конечной точки:", linspace_no_end)  # [0.0, 0.2, 0.4, 0.6, 0.8]

# С возвратом шага
arr, step = np.linspace(0, 1, 5, retstep=True)
print("Массив:", arr)
print("Шаг:", step)  # 0.25

# С указанием типа данных
linspace_int = np.linspace(0, 1, 5, dtype=int)
print("Целочисленные значения:", linspace_int)  # [0, 0, 0, 0, 1]

# Больше элементов
linspace_many = np.linspace(0, 1, 10)
print("10 элементов:", linspace_many)
# [0.0, 0.111..., 0.222..., 0.333..., 0.444..., 0.555..., 0.666..., 0.777..., 0.888..., 1.0]
```

**Функция stats.norm.cdf() - Cumulative Distribution Function**

Что это такое?

CDF (Cumulative Distribution Function) - это функция распределения, которая показывает вероятность того, что случайная величина примет значение меньше или равное заданному значению.

Математическая формула:
Для нормального распределения:
F(x) = P(X ≤ x) = ∫[-∞, x] f(t) dt
где f(t) - функция плотности вероятности нормального распределения.

Практическое использование:

```python
from scipy import stats

# Параметры распределения
mu = 50    # среднее
sigma = 10 # стандартное отклонение

# Вероятность того, что случайная величина ≤ 60
cdf_value = stats.norm.cdf(60, mu, sigma)
print(f"P(X ≤ 60) = {cdf_value:.4f}")  # Пример: 0.8413

# либо сами считаем z:
z_score = (60 - 50)/sigma
prob = stats.norm.cdf(z_score)
print(prob)
```

**Что возвращает функция?**

* Число от 0 до 1, представляющее вероятность

* Например, 0.8413 означает 84.13% вероятность

```python
# Визуализация:

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)  # Функция плотности

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='Плотность вероятности')

# Закрашиваем область под кривой до x=60
x_fill = np.linspace(mu - 4*sigma, 60, 1000)
y_fill = stats.norm.pdf(x_fill, mu, sigma)
plt.fill_between(x_fill, y_fill, alpha=0.5, color='blue')

plt.axvline(60, color='red', linestyle='--', label='x=60')
plt.title(f'Функция распределения: P(X ≤ 60) = {cdf_value:.4f}')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.grid(True)
plt.show()
```

**Функция stats.norm.ppf() - Percent Point Function**

Что это такое?

PPF (Percent Point Function) - это обратная функция распределения (квантильная функция), которая по заданной вероятности возвращает значение, ниже которого случайная величина попадает с этой вероятностью.

Математическая формула:
Если F(x) = p, то x = F⁻¹(p)
где F - функция распределения, p - вероятность.

**Практическое использование:**

```python
# Какое значение соответствует 95-му процентилю?
ppf_value = stats.norm.ppf(0.95, mu, sigma)
print(f"95-й процентиль: {ppf_value:.2f}")  # Пример: 66.45
```

Что возвращает функция?

* Значение случайной величины, соответствующее заданной вероятности

* Например, для p=0.95 возвращает значение, такое что P(X ≤ x) = 0.95

```python
# Визуализация:
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='Плотность вероятности')

# Находим значение для 95-го процентиля
percentile_95 = stats.norm.ppf(0.95, mu, sigma)

# Закрашиваем область под кривой до 95-го процентиля
x_fill = np.linspace(mu - 4*sigma, percentile_95, 1000)
y_fill = stats.norm.pdf(x_fill, mu, sigma)
plt.fill_between(x_fill, y_fill, alpha=0.5, color='blue')

plt.axvline(percentile_95, color='red', linestyle='--', label=f'95-й процентиль: {percentile_95:.2f}')
plt.title('95-й процентиль нормального распределения')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Взаимосвязь между cdf и ppf
# Эти функции являются обратными друг другу:

# Демонстрация обратной связи
p = 0.95
x_value = stats.norm.ppf(p, mu, sigma)  # Из вероятности получаем значение
p_back = stats.norm.cdf(x_value, mu, sigma)  # Из значения возвращаемся к вероятности

print(f"Исходная вероятность: {p:.4f}")
print(f"Соответствующее значение: {x_value:.4f}")
print(f"Вероятность для этого значения: {p_back:.4f}")  # Должно быть равно p
```

**Практические примеры использования**

```python
# Пример 1: Анализ результатов тестирования

# Предположим, результаты теста распределены нормально
test_mean = 75
test_std = 10

# а) Какой процент студентов набрал менее 60 баллов?
p_less_60 = stats.norm.cdf(60, test_mean, test_std)
print(f"Процент студентов с результатом < 60: {p_less_60*100:.2f}%")

# б) Какой балл нужно набрать, чтобы быть в top 10%?
top_10_percentile = stats.norm.ppf(0.90, test_mean, test_std)
print(f"Балл для попадания в top 10%: {top_10_percentile:.2f}")


# Пример 2: Контроль качества

# Длина детали должна быть 100 мм с допуском ±2 мм
length_mean = 100
length_std = 0.8

# а) Какая доля деталей соответствует требованиям?
lower_bound = 98
upper_bound = 102

p_within_tolerance = (stats.norm.cdf(upper_bound, length_mean, length_std) -
                      stats.norm.cdf(lower_bound, length_mean, length_std))
print(f"Доля деталей в допуске: {p_within_tolerance*100:.2f}%")

# б) Найти границы, содержащие 99% деталей
lower_99 = stats.norm.ppf(0.005, length_mean, length_std)  # 0.5% снизу
upper_99 = stats.norm.ppf(0.995, length_mean, length_std)  # 99.5% снизу (или 0.5% сверху)
print(f"99% деталей имеют длину между {lower_99:.2f} и {upper_99:.2f} мм")


# Пример 3: Финансовый анализ

# Доходность инвестиций распределена нормально
return_mean = 0.08  # 8% средняя доходность
return_std = 0.15   # 15% стандартное отклонение

# а) Вероятность отрицательной доходности
p_negative = stats.norm.cdf(0, return_mean, return_std)
print(f"Вероятность отрицательной доходности: {p_negative*100:.2f}%")

# б) Значение, которое с вероятностью 95% будет превышено
value_95 = stats.norm.ppf(0.05, return_mean, return_std)  # 5% снизу = 95% сверху
print(f"С вероятностью 95% доходность превысит: {value_95*100:.2f}%")
```

**Важные замечания**

Параметры по умолчанию: stats.norm.cdf(x) и stats.norm.ppf(p) без указания mu и sigma используют стандартное нормальное распределение (μ=0, σ=1)

Область определения:

cdf: x может быть любым числом

ppf: p должен быть в диапазоне (0, 1)

Точность: Функции используют численные методы для вычисления интегралов и обратных функций, поэтому результаты являются приближенными, но очень точными

Применение: Эти функции широко используются в статистике, машинном обучении, финансах, инженерии и многих других областях

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats  # Предполагается, что эта строка была в предыдущем коде

# Параметры нормального распределения
mu = 50    # среднее значение
sigma = 10 # стандартное отклонение

# Вычисляем вероятность P(X ≤ 60) с помощью функции распределения
cdf_value = stats.norm.cdf(60, mu, sigma)

# Создаем массив x-значений от mu-4σ до mu+4σ (охватывает 99.99% распределения)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Вычисляем значения плотности вероятности для каждого x
y = stats.norm.pdf(x, mu, sigma)  # Функция плотности

# Создаем новую фигуру размером 10x6 дюймов
plt.figure(figsize=(10, 6))

# Рисуем кривую нормального распределения
# 'b-' означает синяя сплошная линия
plt.plot(x, y, 'b-', label='Плотность вероятности')

# Создаем значения для закрашивания области под кривой до x=60
x_fill = np.linspace(mu - 4*sigma, 60, 1000)
y_fill = stats.norm.pdf(x_fill, mu, sigma)

# Закрашиваем область под кривой
# fill_between закрашивает область между кривой и осью x
plt.fill_between(x_fill, y_fill, alpha=0.5, color='blue')

# Добавляем вертикальную линию на уровне x=60
# color='red' - красный цвет, linestyle='--' - пунктирная линия
plt.axvline(60, color='red', linestyle='--', label='x=60')

# Добавляем заголовок с вероятностью P(X ≤ 60)
plt.title(f'Функция распределения: P(X ≤ 60) = {cdf_value:.4f}')

# Подписываем оси
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')

# Добавляем легенду
plt.legend()

# Включаем сетку
plt.grid(True)

# Показываем график
plt.show()
```
